```python
class navigator:
    def __init__(self):
        self.geoResources = {}
        self.shopLocation =(23.799497820002202, 90.35327938170276)
        self.position = 's'
        self.destination = ' '

        for root, dirs, files in os.walk('Navigator/Geojson'):  
            for file in files:
                self.geoResources[file.split('.')[0]] = root+'/'+file

    def changeDestination(self,newDestination):
        self.destination = newDestination
        self.redrawMap()
   

    def drawPathWay(self,shopMap):
      
      def switchPosition(coordinate):
        temp = coordinate[0]
        coordinate[0] = coordinate[1]
        coordinate[1] = temp
        return coordinate

      searchString = self.position + self.destination.split('Shelf')[1]
      with open(self.geoResources[searchString]) as f:
           testWay = json.load(f)

      for feature in testWay['features']:
        path = feature['geometry']['coordinates']

      finalPath = list(map(switchPosition,path))
      folium.plugins.AntPath(finalPath).add_to(shopMap)

    def drawBuilding(self,shopMap):
      shopOutline = self.geoResources[self.destination]
      folium.GeoJson(shopOutline, name="geojson").add_to(shopMap)

    def redrawMap(self):
        shopMap = folium.Map(location = self.shopLocation, width = "100%", zoom_start = 170)
        self.drawPathWay(shopMap)
        self.drawBuilding(shopMap)
        display(shopMap)
```
